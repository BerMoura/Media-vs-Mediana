```python
import numpy as np
from scipy import stats
import pandas as pd

# Dados aleatórios sem outliers:
dados_normais = [10, 12, 14, 16, 18, 20, 12, 14, 21, 10, 12, 20, 24, 18, 10]
# Dados aleatórios com outliers:
dados_com_outliers = [10, 12, 14, 16, 18, 20, 12, 14, 21, 10, 12, 20, 24, 18, 200]

# Cálculos para dados sem outliers:
media_normal = np.mean(dados_normais)
mediana_normal = np.median(dados_normais)

# Cálculos para dados com outliers:
media_outlier = np.mean(dados_com_outliers)
mediana_outlier = np.median(dados_com_outliers)

# Criando a tabela:
tabela = pd.DataFrame({
    "Métrica": ["Média", "Mediana"],
    "Sem Outliers": [media_normal, mediana_normal],
    "Com Outliers": [media_outlier, mediana_outlier]
}).round(1)
print(tabela)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Dados com e sem outliers
dados_normais = [10, 12, 14, 16, 18, 20, 12, 14, 21, 10, 12, 20, 24, 18, 10]
dados_com_outliers = [10, 12, 14, 16, 18, 20, 12, 14, 21, 10, 12, 20, 24, 18, 200]

# Criando o gráfico
plt.figure(figsize=(10, 6))
sns.kdeplot(dados_normais, fill=True, label='Sem Outliers', color='blue', alpha=0.6)
sns.kdeplot(dados_com_outliers, fill=True, label='Com Outliers', color='orange', alpha=0.6)

# Adicionando títulos e legendas
plt.title('Curva de Distribuição dos Dados', fontsize=14)
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.legend(title='Dados', fontsize=10)

# Exibindo o gráfico
plt.show()
```
